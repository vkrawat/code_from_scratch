```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_data.csv')
```

```python
df.head()
```

```python
x=df.YearsExperience
y=df.Salary
```

```python
x=np.array(x)
y=np.array(y)
```

```python
def splits(split):
    #split=0.8
    size=int(df.shape[0]*split)
    xtrain=x[:size]
    ytrain=y[:size]
    xtest=x[size:]
    ytest=y[size:]
    return xtrain,ytrain,xtest,ytest
```

```python
xtrain,ytrain,xtest,ytest=splits(0.8)
```

```python
def gradient(xtrain,ytain,lr):
    a=0
    b=0
    epochs=1000
    n=len(xtrain)
    for i in range(epochs):
        ypred=a+b*xtrain
        cost=1/n*(sum((ytrain-ypred)**2))
        da=-2/n*(sum(ytrain-ypred))
        db=-2/n*(sum(xtrain*(ytrain-ypred)))
        a=a-lr*da
        b=b-lr*db
        print('a={} ,b={} ,cost={} '.format(a,b,cost))
    return a,b;
```

```python
a,b=gradient(xtrain,ytrain,0.04)
```

```python
ypred=a+b*xtrain
```
